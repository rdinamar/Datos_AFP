Se suben los datos a Github en donde se extraen los datos para realizar los graficos de la progresión de los valores de las cuotas de cada Fondo.
Es necesario acotar que la limpieza de los datos se realiza desde GitHub, se borran los puntos de los miles.
Se debe trabajar en limpieza de datos desde la fuente original.
 

```r
library("tidyverse")
library("stringr")
library("readr")
datos <-read.csv("https://raw.githubusercontent.com/rdinamar/Datos_AFP/main/valores_cuota1%202.csv", header = T , sep = ";")


fondo_a<-as.numeric(gsub(",", ".", datos$Fondo.A))
fondo_b<-as.numeric(gsub(",", ".", datos$Fondo.B))
fondo_c<-as.numeric(gsub(",", ".", datos$Fondo.C))
fondo_d<-as.numeric(gsub(",", ".", datos$Fondo.D))
fondo_e<-as.numeric(gsub(",", ".", datos$Fondo.E))
fecha<-as.Date(datos$Día, format = "%d-%m-%y")
fondos<-data.frame(fecha, fondo_a, fondo_b, fondo_c, fondo_d, fondo_e)

head(fondos)
tail(fondos)
summary(fondos)
str(fondos)

library("tidyr")
library("tidyverse")
library("dplyr")
library("ggplot2")

ggplot(fondos, aes(fecha,fondo_a)) +
  geom_line(color = "red")
ggplot(fondos, aes(fecha,fondo_b)) +
  geom_line(color = "orange")
ggplot(fondos, aes(fecha,fondo_c)) +
  geom_line(color = "yellow")
ggplot(fondos, aes(fecha,fondo_d)) +
  geom_line(color = "green")
ggplot(fondos, aes(fecha,fondo_e)) +
  geom_line(color = "blue")


```
